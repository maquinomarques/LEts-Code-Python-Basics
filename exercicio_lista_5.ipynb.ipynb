```python
#1) Faça um programa que receba um dicionario simples de um número associado a uma chave e retorne o dicionario inverso (onde era chave virá valor e vice-versa).

dict_test = {"chave1" : "valor1",
             "chave2" : "valor2",
             "chave3" : "valor3",
             "chave4" : "valor1"}
dict_invertido = {}
lista_chaves = dict_test.keys()
lista_valores = dict_test.values()
for valor in  lista_valores:
  dict_invertido[valor] = [] 
for chave, valor in zip(lista_chaves, lista_valores):
  dict_invertido[valor].append(chave)
```

```python
# 2) Desenvolva uma função que retorne uma lista com todas as chaves de um dicionário que contém o mesmo valor. Por exemplo no dicionário a seguir
#{'a': 3, 'b': 1, 'c': 2, 'd': 3}. A função deve retorna os valores ['a', 'd'] caso o valor procurado seja o 3. 
#E caso não tenha o valor procurado deve-se retornar uma lista vazia.

def procurar_chaves(dict_test, valor_test):
  list_valores = []
  for chaves in dict_test:
    if (dict_test[chaves] == valor_test):
      list_valores.append(chaves)
  return list_valores


```

```python
procurar_chaves(dict_test, "valor1")
```
