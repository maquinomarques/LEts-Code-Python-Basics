```python
#1.1) Desenvolva uma função para calcular a média e o desvio padrão para uma lista de tamanho qualquer (1 ponto) Z=(x−μ)σ
def media_e_desvio_padrao(lista = list):
  soma = 0
  somatorio_desvio_padrao = 0
  for i in lista:
    soma += i
  media = soma/len(lista)
  for i in lista:
    somatorio_desvio_padrao += (media - i)**2 
  desvio_padrao = (somatorio_desvio_padrao/(len(lista)-1))**0.5
  return media, desvio_padrao
```

```python
#1.2) Desenvolva uma função que faça a normalização segundo a formula do Z e retorne a lista de dados normalizados. (1 ponto) Z=(x−μ)/σ

def normalizacao_de_dados(distribuicao_nao_normalizada):
  dados_normalizados =[]
  soma = 0
  somatorio_desvio_padrao = 0
  for i in distribuicao_nao_normalizada:
    soma += i
  media = soma/len(distribuicao_nao_normalizada)

  for i in distribuicao_nao_normalizada:
    somatorio_desvio_padrao += (media - i)**2 
  desvio_padrao = (somatorio_desvio_padrao/(len(distribuicao_nao_normalizada)-1))**0.5

  for i in distribuicao_nao_normalizada:
    z = (i - media)/desvio_padrao
    dados_normalizados.append(z)
  return dados_normalizados

```

```python
#2) Desenvolva uma função que ao receber uma lista de dados quaisquer converta para a escala segundo o MinMaxScaler (2 pontos) S=(x−min)/(max−min)
def scaler_min_max(lista = list):
  min_max_scaler = []
  for i in lista:
    s = ( (i - min(lista)) / (max(lista) - min(lista)))
    min_max_scaler.append(s)
  return min_max_scaler
```

```python
#3) Desenvolva uma função linear_regression que recebe duas listas x e y e devolva os valores dos coeficientes  β0  e  β1  (2 pontos)

def linear_regression(lista_x = list, lista_y = list):
  soma_y = 0
  soma_x = 0
  for x, y in zip(lista_x, lista_y):
    soma_x += x
    soma_y += y
  media_x = soma_x/len(lista_x)
  media_y = soma_y/len(lista_y)

  somatorio_covar_xy = 0
  somatorio_var_x = 0
  for x, y in zip(lista_x, lista_y):
    somatorio_covar_xy += (x - media_x)*(y - media_y)  
    somatorio_var_x += (x - media_x)**2
  beta_1 = somatorio_covar_xy/somatorio_var_x
  beta_0 = media_y - beta_1*media_x

  return beta_0, beta_1
```

```python
#4.2) Desenvolva uma função que receba uma lista de tamanho qualquer e devolva o valor da primeira moda, caso haja uma (1 ponto)
def moda(lista_teste = list):
  valor = 0
  repeticao = 0
  repeticao_i = 0
  lista_ordem = sorted(lista_teste)
  for i in range(len(lista_ordem)):
    if (lista_ordem[i] == lista_ordem[i-1]):
      repeticao_i += 1
      if (repeticao_i > repeticao):
        repeticao = repeticao_i
        valor = lista_ordem[i]
      else:
        continue
    else:
      repeticao_i = 0
  return valor, repeticao
```
