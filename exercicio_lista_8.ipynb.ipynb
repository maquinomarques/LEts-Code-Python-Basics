```python
#1) Crie uma função chamada formulario, onde o objetivo dela é pedir essas informações como inputs do Usuário (a função não vai receber parâmetros, apenas os inputs) (1 ponto),
def formulario():
  nome = input("Nome completo do funcionário: ")
  idade = int(input("Sua idade: "))
  estado = input("Estado em que reside: ")
  setor_de_atuação = input("Qual seu setor de atuação: ")
  tempo_de_empresa = int(input("Quantos anos completos tem de empresa: "))
  salario = int(input("Qual seu salário: "))
  return nome, idade, estado, setor_de_atuação, tempo_de_empresa, salario
```

```python
#2) Após criar a função formulario, crie uma função chamada de armazenamento onde essa função chamada a função formulario para receber os inputs e depois armazena essas 
#   informações em um dicionário pré-criado (este dicionário deve servir de parâmetro para a função armazenamento) (1 ponto)

livro_de_registro = {}

livro_de_registro["nome"] = []
livro_de_registro["idade"] = []
livro_de_registro["estado"] = []
livro_de_registro["setor da empresa"] = []
livro_de_registro["tempo de empresa"] = []
livro_de_registro["salario"] = []
repetir = "sim"
while (repetir.lower() == "sim"):
  lista = formulario()
  livro_de_registro["nome"].append(lista(0))
  livro_de_registro["idade"].append(lista(1))
  livro_de_registro["estado"].append(lista(2))
  livro_de_registro["setor da empresa"].append(lista(3))
  livro_de_registro["tempo de empresa"].append(lista(4))
  livro_de_registro["salario"].append(lista(5))
  repetir = input("Deseja repetir? (sim/não")
print(livro_de_registro)
```
