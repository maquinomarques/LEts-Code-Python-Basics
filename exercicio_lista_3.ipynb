```python
#1.1
cnt = 0
numero = int(input("Digite um número "))
fatorial = 1
if (numero == 0):
  print(f"Fatorial é {fatorial}")
elif (numero < 0):
  print("Número inválido")
else:
  cnt += 1
  while (cnt <= numero):
    fatorial *= cnt
    cnt += 1
print(f'O fatorial é {fatorial}')
```

```python
#1.2
numero = int(input("Digite um número para o calculo do fatorial "))
fatorial = 1

for i in range(numero, 0, -1 ):
  fatorial *= i

print(f"O fatorial de {numero} é {fatorial}")
```

```python
#2) Desenvolva um código que onde pede para o usuário que digite um número inteiro entre 1 e 200, e o código deve printar todos os números primos 
#entre 2 e o número digitado. Número primo é um que seja divisível apenas por 1 ou ele mesmo. Exemplo de números primos : 2, 3, 5, 7, 11, 13, 17 e assim por diante.
#2.1) Desenvolva o código na versão while (1 ponto)
cnt = 0
memoria_auxiliar = 0
numero = int(input("Digite um número para saber quais são os primos até ele "))
primos = 0
while (cnt < numero):
  cnt += 1 
  divisores = 0
  memoria_auxiliar = 0
  while (memoria_auxiliar <= cnt):
    memoria_auxiliar += 1
    if (cnt % memoria_auxiliar == 0):
      divisores += 1
    else:
      continue
  if (divisores == 2):
    print(f"{cnt} é primo")
    primos += 1

print(f"Existem {primos} primos entre 2 e {numero}")
```

```python
#2.2) Desenvolva o código na versão for 

primos = 0
numero = int(input("Digite um número para saber quais são os primos até ele "))
for i in range (1, numero + 1):
  divisores = 0
  for a in range(1,i+1):
    if (i%a == 0):
      divisores += 1
    else: 
      continue
  if (divisores == 2):
    print(f"{i} é primo ")
    primos += 1
print(f"Existem {primos} primos entre 2 e {numero}")
```
