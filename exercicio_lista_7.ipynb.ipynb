```python
#3) Desenvolver uma função nomeada nota_mais_frequente, onde será recebido uma lista com 40 notas de alunos e a função deve retornar uma lista com respectivamente 
#   a indicação de qual é a nota mais frequente e o valor de quantas vezes ocorreu essa nota.

def moda(lista_teste = list):
  nota = 0
  repeticao = 0
  repeticao_i = 0
  lista_ordem = sorted(lista_teste)
  for i in range(len(lista_ordem)):
    if (lista_ordem[i] == lista_ordem[i-1]):
      repeticao_i += 1
      if (repeticao_i > repeticao):
        repeticao = repeticao_i
        nota = lista_ordem[i]
      else:
        continue
    else:
      repeticao_i = 0
  return nota, repeticao +1
```

```python
#4) Criar uma função nomeada palindromo, onde a função recebe uma lista de palavras e a saída será a quantidade de palindromos dentro da lista.
def qtd_palindromos(lista_de_palavras:list):
  qtd_palindromos = 0
  for palavra in lista_de_palavras:
    palavra_invertida = ""
    for caracter in palavra[::-1]:
      palavra_invertida += caracter
    if palavra == palavra_invertida:
      qtd_palindromos += 1
  return qtd_palindromos
#def palindromos(lista_palavras):
# cnt = 0
# for word in lista_palavra:
#   if word == word[::-1]
#     cnt += 1
# return cnt
```

```python
#5) O objetivo deste exercício é desenvolver uma função nomeada como conversor_numero, 
#onde irá receber um valor númerico inteiro na base decimal e converter para um valor na base binária.

numero = 10
def conversor_numero(numero:int):
  lista_binario = []
  dividendo = numero
  numero_binario =""
  while (dividendo > 1):
    resto = dividendo%2
    dividendo //= 2
    lista_binario.append(resto)
  if (dividendo != 0):
    lista_binario.append(dividendo)

  for i in lista_binario[::-1]:
    numero_binario += str(i)

  return numero_binario
```
